### Q11) Multiple Files

There are 5 files within the `gapminder`. Use a loop to create the dataframe `region_df` that contains a row for each region/continent that has a average for each of the gdpPerCap columns. The region/continent needs to be extracted from the filename.

```python
import os
```

```python
folder_path = 'data/gapminder/' #rel path
region_df = pd.DataFrame() #empty df

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        region = os.path.splitext(filename)[0].split('_')[2] # Extract the region/continent from the filename
        
        file_path = os.path.join(folder_path, filename) #full file path
        
        # Read the CSV file and append it to the region_df DataFrame
        df = pd.read_csv(file_path)
        df['Region'] = region  # Add Region column with the region/continent name
        region_df = pd.concat([region_df, df], ignore_index=True)

region_df.head(100)
```

```python
columns = region_df.columns.tolist() #list of columns

column = 'Region' #regions column with extracted regions/continent from filename
new_index = 0 # where the old continent column is indexed (position)

columns.remove('Region')#removes regions column from its index
columns.insert(new_index, column)
region_df = region_df.reindex(columns=columns)
region_df = region_df.drop(['continent'], axis=1)
region_df.head(100)

```

### Q12) Plotting regional averages.
Create a line plot that showes how the average GDP per capita for each region has changed over time.

```python
avg_gdp = region_df.groupby('Region').mean()
avg_gdp = avg_gdp.T
avg_gdp
```

```python
import matplotlib.pyplot as plt

region_data = {
    'Year': avg_gdp.index,
    'Africa': avg_gdp['africa'],
    'Americas': avg_gdp['americas'],
    'Asia': avg_gdp['asia'],
    'Europe': avg_gdp['europe'],
    'Oceania': avg_gdp['oceania']
}
region_avg_df = pd.DataFrame(region_data)

plt.figure(figsize=(20, 5)) 

plt.plot(region_avg_df['Year'], region_avg_df['Africa'], label='Africa')
plt.plot(region_avg_df['Year'], region_avg_df['Americas'], label='Americas')
plt.plot(region_avg_df['Year'], region_avg_df['Asia'], label='Asia')
plt.plot(region_avg_df['Year'], region_avg_df['Europe'], label='Europe')
plt.plot(region_avg_df['Year'], region_avg_df['Oceania'], label='Oceania')

plt.title('Average GDP per Capita by Region')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')

plt.legend()

plt.show()
```
